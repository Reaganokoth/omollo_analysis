```python
AGB_vs_Habitat = smf.ols(formula='Average_AGB ~ C(Habitat)', data=data_diani).fit()
summary = AGB_vs_Habitat.summary()
print(summary)
```

```python
BGB_vs_Habitat = smf.ols(formula='Average_BGB ~ C(Habitat)', data=data_diani).fit()
summary = BGB_vs_Habitat.summary()
print(summary)
```
